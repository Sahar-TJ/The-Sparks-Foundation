__Decision Trees__ are a type of Supervised Machine Learning that is you explain what the input is and what the corresponding output is in the training data, where the data is continuously split according to a certain parameter.

### Importing Libraries and Iris Dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import datasets
```

```python
# Loading iris dataset
df = pd.read_csv('Iris.csv')
df = df.drop(df.columns[[0]], axis=1) #removing the Id coloumn from the dataset
df.head()
```

```python
#statistics of the dataset 
df.describe()
```

```python
#Summary of Dataset
df.info()
```

### Relationships between columns of the dataset

```python
#Checking the relationship
sns.pairplot(df, hue='Species')
```

### Correlation between columns by using Heatmap

```python
plt.figure(figsize=(10,11))
sns.heatmap(df.corr(),annot=True)
plt.plot()
```

__We can see Petal Length and Petal Width are almost correlatted.__

### Using  boxplot to see how the “Species” is distributed with other columns

```python
#boxplot
plt.figure(figsize=(12,10))
plt.subplot(2,2,1)
sns.boxplot(x="Species",y="SepalLengthCm",data=df)
plt.subplot(2,2,2)
sns.boxplot(x="Species",y="SepalWidthCm",data=df)
plt.subplot(2,2,3)
sns.boxplot(x="Species",y="PetalLengthCm",data=df)
plt.subplot(2,2,4)
sns.boxplot(x="Species",y="PetalWidthCm",data=df)
```

### Splitting the Dataset

```python
# Splitting the dataset
x = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].values
y = df['Species'].values

(train_x, test_x, train_y, test_y) = train_test_split(x, y, train_size=0.7, random_state=1)
```

### Classification and finding the Accuracy 

```python
#Classification
dtc = DecisionTreeClassifier()
dtc.fit(train_x, train_y)
#Predict response for test dataset
y_pred_tree=dtc.predict(test_x)
print(y_pred_tree)
#Accuracy
print("\nAccuracy: ",dtc.score(test_x, test_y))
```

#### The accuracy of the model is 95.55% which is quite good.

### Plotting the Graph

```python
from sklearn import tree
fig,ax=plt.subplots(figsize=(15,8))
tree.plot_tree(dtc,feature_names=list(df),class_names='Species',filled=True)
```
